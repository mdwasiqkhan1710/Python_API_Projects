```python
response = requests.get(url, headers=headers)
print(response)
```

```python
print("Status Code:", response.status_code)
print("Content-Type:", response.headers.get("Content-Type"))
print("Response Text:")
print(response.text)
```

```python
payload = response.json()

## Printing the response which we are getting from API
print(payload)
```

## ------------------------------Transforming Data----------------------------

```python
formatted_response = json.dumps(payload, indent=4)
print(formatted_response)
```

```python
standings_list = payload['rankings']
standings_list
```

```python
# Formatting the standings list for better understanding
formatted_standings_list = json.dumps(standings_list, indent=4)
print(formatted_standings_list)
```

```python
## Points for each team in Premier League are calculated as follows -
## 3 points for a win, 1 point for a draw, and 0 points for a loss.

Manchester_City_Games_Won = int(standings_list[0]['won'])
Manchester_City_Games_Drawn = int(standings_list[0]['tied'])

Manchester_City_Points = (Manchester_City_Games_Won * 3) + (Manchester_City_Games_Drawn * 1)

print(Manchester_City_Points)
```

```python
## Setting up rows and columns for dataframe to store data in clean table like format 
rows = []
column_names = ['season', 'position', 'team', 'played', 'won', 'draw', 'lost', 'goals_for', 'goals_against', 'goals_diff', 'points']

## Looping through all the teams in standings list to create the dataframe
for club in standings_list:
    season = 2024
    position = int(club['team_rank'])
    team = club['team_name']
    played = int(club['games_played'])
    won = int(club['won'])
    draw = int(club['tied'])
    lost = int(club['lost'])
    goals_for = int(club['goals_scored'])
    goals_against = int(club['goals_against'])
    goals_diff = int(goals_for) - int(goals_against)

    points = int(won) * 3 + int(draw) * 1

    ## Created tuple for all the records of each team/club
    tuple_of_club_records = (season, position, team, played, won, draw, lost, goals_for, goals_against, goals_diff, points)

    ## Appending the tuple in the rows
    rows.append(tuple_of_club_records)
```

```python
## Checking if for loop worked and all data has been successfully gathered
print(rows)
```

```python
## Creating the final dataframe
df = pd.DataFrame(rows, columns=column_names)
```

```python
db_connection = connector.connect(
    host = MYSQL_HOST,
    port = MYSQL_PORT,
    user = MYSQL_USER,
    password = MYSQL_PASSWORD,
    database = MYSQL_DATABASE
)

cur = db_connection.cursor()
print(f"[SUCCESS] - Successfully connected to database!")
```

```python
sql_table = "standings"
cur.execute("SHOW TABLES LIKE %s", (f"{sql_table}",))

if cur.fetchone() is None:
    raise SystemExit(f"Table with name - {sql_table} does not exist...please create it!")
else:
    print(f"[Success] - {sql_table} exists...Continue!")
```

```python
user_team = "LIVERPOOL'; DROP TABLE standings"
cur.execute("SELECT * FROM standings WHERE team = %s", (user_team,))
```

```python
# Starting the UPSERT Operations

table_cols = ['season', 'position', 'team', 'played', 'won', 'draw', 'lost', 'goals_for', 'goals_against', 'goals_diff', 'points']

standings_df = df[table_cols]
```

```python
standings_df
```

```python
standings_records_tuples = standings_df.itertuples(index=False, name=None)

list_of_standings_records_tuples = list(standings_records_tuples)
```

```python
UPSERT_SQL = f"""
    INSERT INTO {sql_table}
    (season, position, team, played, won, draw, lost, goals_for, goals_against, goals_diff, points)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s) AS src
    ON DUPLICATE KEY UPDATE
    position = src.position,
    team = src.team,
    played = src.played,
    won = src.won,
    draw = src.draw,
    lost = src.lost,
    goals_for = src.goals_for,
    goals_against = src.goals_against,
    goals_diff = src.goals_diff,
    points = src.points;
"""
```

```python
no_of_rows_uploaded_to_mysql = len(list_of_standings_records_tuples)
```

```python
try:
    cur.executemany(UPSERT_SQL, list_of_standings_records_tuples)
    db_connection.commit()
    print(f"[Success] - UPSERT attempted for {no_of_rows_uploaded_to_mysql} rows")
except Exception as e:
    db_connection.rollback()
    print(f"[Error] - Rolled back due to this error.......: \n{e}")
finally:
    cur.close()
    db_connection.close()
    print("All databse connections now closed...\n\nClean up completed!")
```
